```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('example.db')
cursor = conn.cursor()
```

```python
cursor.execute('''CREATE TABLE IF NOT EXISTS users(user_id INTEGER PRIMARY KEY, name TEXT NOT NULL,email TEXT NOT NULL UNIQUE)''')
```

```python
cursor.execute ("INSERT OR IGNORE into users(user_id,name,email) VALUES (1,'Alice','alice@example.com')")
cursor.execute("INSERT OR IGNORE into users(user_id,name,email) VALUES (2, 'Bob', 'bob@example.com')")
conn.commit()
print("Databased and table created successfully.")
```

```python
users_df = pd.read_sql_query("SELECT * FROM users", conn)
display(users_df)
```

```python
conn.close()
```
